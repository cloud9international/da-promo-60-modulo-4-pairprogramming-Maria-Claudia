```python
import pandas as pd
```

```python


# 1. Cargar el archivo CSV
df = pd.read_csv('friends_quotes.csv')
df
```

```python
# 2. Convertir la columna (reemplaza 'nombre_columna' por el tuyo)
# Usamos 'Int64' (con mayúscula) si el archivo tiene valores vacíos (NaN)
# O .astype(int) si estás seguro de que no hay celdas vacías
df['episode_number'] = df['episode_number'].astype(int)
df['season'] = df['season'].astype(int)

# 3. Guardar el resultado en un nuevo CSV
df.to_csv('archivo_procesado.csv', index=False)

print("Conversión completada con éxito.")
```
